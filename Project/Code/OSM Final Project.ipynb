```python
def assess_result(node,result):
    if result == True:
        if node['Updates']:            
            node['Updates'] += 1 
        else:
            node['Updates'] = 1

def update_name(name, mapping):
    old_name = name
    if mapping:
        for k,v in mapping.iteritems():
            name = name.strip()
            name = re.sub(r'(\W)',' ',name)
            kv = re.compile(r'\s'+k+'$',flags = re.I)
            kv = kv.search(name.lower())
            vv = re.compile(v,flags = re.I)
            vv = vv.search(name.lower())
            if kv != None and vv == None:
                name = name.replace(name[kv.start(0):kv.end(0)+1],v)
            else:
                continue
    name = name.title()
            
    if name != old_name:
        return name,True
    else:
        return name,False

def update_phone(phone):
    old_phone = phone
    phone = re.sub(r'\s|[)]|[(]|[+]|(^0)','',phone,flags = re.M)
    phone = re.sub(r'(^27)','',phone)
    phone = re.sub(r'(^0)','',phone)
    
    if len(phone) == 9:
        phone = '{0} {1} {2} {3}'.format('+27',phone[:2],phone[2:5],phone[5:])
    else:
        phone = 'skip'
        
    if old_phone != phone and phone != 'skip':
        return (phone,True)
    else:
        return (phone,False)
        
    
def shape_element(element):
    node = {}
    k = ''
    v = ''
    lat = 0.0
    lon = 0.0
    
    if element.tag == "node" or element.tag == "way" :
        for c in element.iter("node"):
            node['type'] = 'node'
            for k,v in c.attrib.iteritems():
                if  k in CREATED:
                    if not 'created' in node.keys():
                        node['created'] = dict()
                    node['created'][k] = v
                elif k == 'lon':
                    lon = float(v)
                elif k == 'lat':
                    lat = float(v)
                    node['pos']=[lat,lon]
                else:
                    node[k] = v
                    
        for e in element.iter("way"):
            for k,v in e.attrib.iteritems():
                node['type'] = 'way'
                if  k in CREATED:
                    if not 'created' in node.keys():
                        node['created'] = dict()
                    node['created'][k] = v
                    
        for e in element.iter("nd"):
            ref = e.attrib["ref"]
            if not 'node_refs' in node.keys():
                node['node_refs'] = []
            node['node_refs'].append(ref)    
            
        for e in element.iter("tag"):
            k = e.attrib["k"]
            v = e.attrib["v"]
            result = False
            if not problemchars.search(k):
                if k[:4] == 'addr':
                    result = False     
                    if not 'address' in node.keys():
                        node['address'] = dict()
                    address_key = k.split(':')
                    if len(address_key) <= 2 and len(address_key) > 1:
                        v,result = update_name(v,street_map)
                        node['address'][address_key[1]] = [v,result]
                else:
                    result = False
                    if k == 'phone':
                        v,result = update_phone(v)
                    elif k in ['name']:
                        v,result = update_name(v,'')
                    node[k] = [v,result] 
                
        return node
    else:
        return None

def insert_data(dbase,doc,data):
    client = pym.MongoClient()
    c = pym.Connection()
    c.drop_database(dbase)
    db = client[''+dbase+''][''+doc+'']
    db.insert(data)
    print db.find_one()

def process_map(doc,path, pretty = False):
    file_in = "{0}{1}{2}".format(path,'//data//raw//',doc)
    file_out = "{0}{1}{2}.json".format(path,'//data//clean//',doc)
    data = []
    with open(file_out,"w") as out:
        for _, element in ET.iterparse(file_in):
            el = shape_element(element)
            if el:
                data.append(el)
        json.dump(data,out)
    return data


```

```python
def unpack(o_id,k_0,v_0):
    if type(v_0) != bool:
        if not type(v_0) in [str,unicode]:
            if type(v_0) == list:
                for v_1 in v_0:
                    unpack(o_id,k_0,v_1)
            elif type(v_0) == dict:
                for k_1,v_1 in v_0.iteritems()
                    unpack(o_id,k_1,v_1)          
        else:
            count_values(o_id,k_0,v_0)
            check_char(o_id,'',k_0)
            check_char(o_id,k_0,v_0)
```
